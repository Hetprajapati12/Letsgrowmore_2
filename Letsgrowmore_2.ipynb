**TASK 1-Image to Pencil Sketch with Python**

*Importing required libraries*

```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
```

*Loading the image*

```python
img = cv2.imread('Het.jpg', 1)
```

```python
cv2_imshow(img)
```

*Converting an image into gray_scale image*

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(img)
```

*Inverting the image*

```python
img_invert = cv2.bitwise_not(img_gray)
```

```python
cv2_imshow(img_invert)
```

*Smoothing the image*

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
```

```python
cv2_imshow(img_smoothing)
```

*Obtaining the final sketch*

```python
def dodgeV2(x, y):

    return cv2.divide(x, 255 - y, scale=256)
```

*Final Output as Pencil Sketch*

```python
final_img = dodgeV2(img_gray, img_smoothing)

cv2_imshow(final_img)

```
